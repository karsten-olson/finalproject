```python
bos_pd = pd_new.loc[pd_new['Team'] == 'BOS']
cle_pd = pd_new.loc[pd_new['Team'] == 'CLE']
gsw_pd = pd_new.loc[pd_new['Team'] == 'GSW']
hou_pd = pd_new.loc[pd_new['Team'] == 'HOU']
nop_pd = pd_new.loc[pd_new['Team'] == 'NOP']
phi_pd = pd_new.loc[pd_new['Team'] == 'PHI']
tor_pd = pd_new.loc[pd_new['Team'] == 'TOR']
uta_pd = pd_new.loc[pd_new['Team'] == 'UTA']
```

```python
 %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = cle_pd.TotalNumberPossession.values.reshape(-1, 1)
y = cle_pd.FavDelta.values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
X
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=False)
```

```python
print('Weight coefficients: ', model.coef_)
print('y-axis intercept: ', model.intercept_)
```

```python
x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
plt.scatter(X, y, c='Maroon')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
plt.xlabel("Finals Total Number Possession")
plt.ylabel("Finals Net Rating")
plt.axvline(x= 93.3)
plt.savefig("Data Fig/CLE_Finals_Net_PPP.png")
```
